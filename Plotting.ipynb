```python
def P_ig( V,T ):
    R = 8.314       # ideal gas constant, joules / kelvin
    P = R * T / V
    return P
```

```python
def P_vdw( V,T,T_c,P_c ):
    R = 8.314       # ideal gas constant, joules / kelvin
    a = (27./64) * (R*T_c)**2 / P_c
    b = R * T_c / (8. * P_c)
    P = R * T / ( V - b ) - a / V ** 2
    return P
```

```python
def P_rk(V, T, T_c, P_c):
    R = 8.314 # ideal gas constant, joules / kelvin
    a = 0.42748 * R**2 * T_c**2.5 / P_c
    b = 0.08662 * R * T_c / P_c
    P = R * T / ( V - b ) - a / ( T ** 0.5 * V * ( V - b ) )
    return P
```

```python
temps = [ 273.15,298.15,323.15,348.15,373.15,398.15 ]
      # data contains T_c,P_c tuples
data = { 'H2' : ( 33.19,1313000 ),
         'O2' : ( 154.6,5043000 ),
         'N2' : ( 126.2,3400000 ),
         'CO2': ( 304.2,7383000 ),
         'Ar' : ( 150.9,4894000 ),
         'C4H10': ( 408.1,3648000 ) }
```

```python
import matplotlib.pyplot as plt
volume = 0.044 # m**3

Pcyl_ig  = []
Pcyl_vdw = []
Pcyl_rk  = []
for temp in temps:
    Pcyl_ig.append( P_ig( volume,temp ) )
    Pcyl_vdw.append( P_vdw( volume,temp,data[ 'Ar' ][ 0 ],data[ 'Ar' ][ 1 ] ) )
    Pcyl_rk.append( P_rk(  volume,temp,data[ 'Ar' ][ 0 ],data[ 'Ar' ][ 1 ] ) )

plt.plot( temps,Pcyl_ig,'g-',linewidth=2,label='Ideal Gas' )
plt.plot( temps,Pcyl_vdw,'b-',linewidth=2,label='van der Waals' )
plt.plot( temps,Pcyl_rk,'r-',linewidth=2,label='Redlich–Kwong' )
plt.legend()
plt.show()
```
